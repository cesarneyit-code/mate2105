## Prueba local rápida
Después de completar las funciones, esta celda debe producir probabilidades por fila.

```python
def train_for_evidence(
    X_train,
    y_train,
    X_validation,
    y_validation,
    *,
    n_hidden=32,
    learning_rate=0.15,
    epochs=120,
    l2=1e-4,
    seed=7,
):
    n_classes = int(np.max(y_train)) + 1
    Y_train = one_hot(y_train, n_classes)
    params = init_parameters(X_train.shape[1], n_hidden, n_classes, seed=seed)
    rows = []
    for epoch in range(epochs + 1):
        P, cache = forward_pass(X_train, params)
        loss = cross_entropy_loss(P, Y_train)
        loss += 0.5 * l2 * (
            np.sum(params["W1"] ** 2) + np.sum(params["W2"] ** 2)
        )
        if epoch % 10 == 0 or epoch == epochs:
            rows.append(
                {
                    "epoch": epoch,
                    "loss": float(loss),
                    "train_accuracy": float(
                        accuracy_score(y_train, predict(X_train, params))
                    ),
                    "validation_accuracy": float(
                        accuracy_score(y_validation, predict(X_validation, params))
                    ),
                }
            )
        if epoch < epochs:
            grads = backward_pass(X_train, Y_train, cache, params, l2=l2)
            params = update_parameters(params, grads, learning_rate)
    return params, pd.DataFrame(rows)


digits = load_digits()
X = digits.data / 16.0
y = digits.target
X_train, X_validation, y_train, y_validation = train_test_split(
    X, y, test_size=0.25, stratify=y, random_state=42
)
params, history = train_for_evidence(
    X_train[:700],
    y_train[:700],
    X_validation[:250],
    y_validation[:250],
)
evidence_summary = {
    "initial_loss": float(history.iloc[0]["loss"]),
    "final_loss": float(history.iloc[-1]["loss"]),
    "validation_accuracy": float(history.iloc[-1]["validation_accuracy"]),
    "random_baseline_accuracy": 0.10,
}
print(evidence_summary)
history
```
