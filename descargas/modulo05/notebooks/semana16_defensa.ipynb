```python
def puntaje_total(rubrica):
    valores = pd.to_numeric(rubrica["puntaje"], errors="coerce")
    return valores.sum(skipna=True)

# Ejemplo de uso después de evaluar
rubrica["puntaje"] = [2, 2, 1, 2, 1]
puntaje_total(rubrica)
```
