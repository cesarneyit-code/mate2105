```python
evidencia = pd.DataFrame([
    {"bloque": "pregunta", "estado": "pendiente", "evidencia": "frase evaluable"},
    {"bloque": "baseline", "estado": "pendiente", "evidencia": "métrica baseline"},
    {"bloque": "modelo", "estado": "pendiente", "evidencia": "pipeline y parámetros"},
    {"bloque": "validación", "estado": "pendiente", "evidencia": "CV/dev/test"},
    {"bloque": "errores", "estado": "pendiente", "evidencia": "matriz, slices o casos"},
    {"bloque": "reproducibilidad", "estado": "pendiente", "evidencia": "requirements + manifest"},
])
evidencia
```

```python
def completion_score(frame):
    frame = frame.copy()
    return float((frame["estado"].str.lower() == "listo").mean())

# Cambia estados a "listo" durante la clínica.
evidencia.loc[[0, 1], "estado"] = "listo"
completion_score(evidencia)
```

```python
claims = pd.DataFrame([
    {"afirmacion": "supera baseline", "artefacto": "tabla_modelos.csv", "limite": "partición interna"},
    {"afirmacion": "slice crítico", "artefacto": "errores_slices.csv", "limite": "n pequeño"},
])
claims
```

```python
issues = pd.DataFrame([
    {"control": "hash de datos", "estado": "pendiente", "severidad": "crítica", "responsable": "grupo"},
    {"control": "semilla", "estado": "pendiente", "severidad": "alta", "responsable": "grupo"},
    {"control": "tipografía", "estado": "pendiente", "severidad": "baja", "responsable": "grupo"},
])
severity_order = pd.CategoricalDtype(["crítica", "alta", "media", "baja"], ordered=True)
issues["severidad"] = issues["severidad"].astype(severity_order)
issues.sort_values("severidad")
```

```python
audit_summary = {
    "completion": completion_score(evidencia),
    "critical_open": int(((issues["severidad"] == "crítica") & (issues["estado"] != "listo")).sum()),
    "claims_with_support": int((claims["artefacto"].str.len() > 0).sum()),
}
audit_summary
```
