## Prueba local sugerida

```python
data = load_breast_cancer(as_frame=True)
X = data.data
y = data.target
X_train, X_dev, X_test, y_train, y_dev, y_test = make_splits(
    X, y, random_state=42
)

candidates = {
    "baseline": build_baseline(),
    "logreg_C_0_1": build_pipeline(C=0.1),
    "logreg_C_1": build_pipeline(C=1.0),
}
cv_results = {
    name: cross_validate_model(model, X_train, y_train, cv=3, scoring="f1")
    for name, model in candidates.items()
}
best_name, conservative_score = choose_best_model(cv_results)
selected_model = candidates[best_name]
selected_model.fit(X_train, y_train)
y_dev_pred = selected_model.predict(X_dev)
dev_metrics = metric_report(y_dev, y_dev_pred)
slice_report = slice_error_report(
    X_dev,
    y_dev,
    y_dev_pred,
    feature="mean radius",
    threshold=float(X_dev["mean radius"].median()),
)
evidence_summary = {
    "selected_model": best_name,
    "conservative_cv_score": float(conservative_score),
    "baseline_cv_f1": float(cv_results["baseline"]["mean"]),
    "development_f1": float(dev_metrics["f1"]),
    "development_errors": int(np.sum(y_dev_pred != np.asarray(y_dev))),
}
print(evidence_summary)
slice_report
```
